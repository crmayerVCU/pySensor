```python
def closest2(dist_matrix,index,number):
    vect=dist_matrix[index].copy()
    closest=np.zeros([number,2])
    for i in range (number):
        minimum=np.min(vect)
        closest[i,0]=np.where(vect==minimum)[0][0]
        closest[i,1]=minimum
        vect[closest[i,0]]=1e20
    return(closest)
```

```python
def recommend_new(User,Data_df,user_dist_matrix,N,P):
    Data_df=Data_df.copy()
    user_dist_matrix=user_dist_matrix.copy()
    User_ind=np.where(Data_df.columns==User)[0][0]
    print(User_ind)
    #print(user_dist_matrix.shape,User_ind,N)
    neighbors=closest2(user_dist_matrix,User_ind,N)
    #print(neighbors.shape)
    #print(neighbors)
    neighbor_dat=np.zeros([N,Data_df.values.shape[0]])
    for n in range(N-1):
        neighbor_dat[n,:]=Data_df.values[:,neighbors[n,0]]
    neighbor_ave=np.nanmean(np.nan_to_num(neighbor_dat),axis=0)
    new_page_mask=np.nan_to_num(Data_df.values[:,User_ind])==0
    #print(neighbor_ave)
    X=neighbor_ave*new_page_mask.T
    print(X.shape)
    #print(X)
    rec_df=pd.DataFrame({'PageID':Data_df.index,
                         'PageIndex':range(X.shape[0]),
                         'Recommendation Strength':X})
    print(rec_df.sort_values(by='Recommendation Strength',ascending=False)[0:P])
```

```python
def recommend_weighted(User,Data_df,user_dist_matrix,P):
    Data_df=Data_df.copy()
    user_dist_matrix=user_dist_matrix.copy()
    User_ind=np.where(Data_df.columns==User)[0][0]
    print(User_ind)
    N=len(Data_df.columns)
    print(N)
    #Closest Users and distances between users to query
    neighbors=closest2(user_dist_matrix,User_ind,N)
    
    #store User Pages 
    neighbor_dat=np.zeros([N,Data_df.values.shape[0]])
    for n in range(N-1):
        #stores User Pages in the order of neighbors
        neighbor_dat[n,:]=Data_df.values[:,neighbors[n,0]]
    
    #Gives weighted average of Page Vectors from N Users that were closest to query
    neighbor_ave=np.average(np.nan_to_num(neighbor_dat),axis=0,weights=1/(1+neighbors[:,1]))
    
    #Find where user has not viewed pages
    new_page_mask=np.nan_to_num(Data_df.values[:,User_ind])==0
    
    #Returns weights for unviewed pages, all other pages forced to zero weighting
    X=neighbor_ave*new_page_mask.T
    
    
    #output recommendations by strength, where strength is the magnitude of the difference between weighted average 
    #user to query 
    rec_df=pd.DataFrame({'PageID':Data_df.index,
                         'PageIndex':range(X.shape[0]),
                         'Recommendation Strength':X})
    print(rec_df.sort_values(by='Recommendation Strength',ascending=False)[0:P])
```

```python
recommend_weighted('user482',Data,Users,10)
```
