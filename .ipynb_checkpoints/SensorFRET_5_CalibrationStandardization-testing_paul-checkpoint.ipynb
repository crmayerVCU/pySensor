# All Image Data is Read into Calibration Object
All spectral images, Masks1, and Masks2 data is read into the tif_array attribute of FRET_Calib. 

```python
self=C
interpolation=True
xdim=100
ydim=100

if interpolation==True:
    #imresize(arr, size, interp='bilinear', mode=None)
    #nested_tiflist[exp groups][F1,F2,Masks][files]
    nested_tiflist=self.file_names
    #get size of the first image for array containers
    channels_i,x,y=tf.TiffFile(nested_tiflist[0][0][0]).asarray().shape

    x=int(xdim)
    y=int(ydim)

    # print("channels,x,y",channels_i,x,y)

    if self.channels==[]:        
        self.channels=channels_i        
    if channels_i!=self.channels:
        print('Number of Channels in Image Do Not Match Specified Emission Channels. Could be due to extra transmission channel or incorrect specification of the number of emission channels. This will default to the specified number of emission channel')
    self.img_num=[]        
    self.X=x
    self.Y=y
    tif_container=[]
    group_num=[]        
    dim3=[]        
    for group in tnrange(len(nested_tiflist)):
        #determines number of images in subdir group
        im_count=len(nested_tiflist[group][0])
        self.img_num.append(im_count)
        #creates ndarray to store all images
        F1=np.zeros((im_count,self.channels,x,y))
        F2=np.zeros((im_count,self.channels,x,y))
        Mask1=np.ones((im_count,x,y))
        Mask2=np.ones((im_count,x,y))
        image_num=[]            
        dim2=[]            
        for image in range(im_count):#loops through subdir list and groups all images into array container
            if fn_print==True:
                print(nested_tiflist[group][0][image])
                print(nested_tiflist[group][1][image])
                print(nested_tiflist[group][2][image])
                print(nested_tiflist[group][3][image])
            F1t=tf.TiffFile(nested_tiflist[group][0][image]).asarray()[:self.channels,:,:]
            F2t=tf.TiffFile(nested_tiflist[group][1][image]).asarray()[:self.channels,:,:]
            for channel in range(self.channels):
                F1[image,channel,:,:]=imresize(F1t[channel,:,:],size=(x,y),interp='bilinear')
                F2[image,channel,:,:]=imresize(F2t[channel,:,:],size=(x,y),interp='bilinear')

            if len(nested_tiflist[group][2])>1:
                tmask=tf.TiffFile(nested_tiflist[group][2][image]).asarray()
                tmasknew=imresize(tmask,size=(x,y),interp="bilinear")
                Mask1[image,:,:]=label(tmasknew)
                #tlabel=label(tf.TiffFile(nested_tiflist[group][2][image]).asarray())
                
                #Mask1[image,:,:]=imresize(tlabel,size=(x,y),interp='bilinear')

            #if Masks2 directory is empty, then no directory will be read in get_nested_subdirectories(self). If there
            #are images in Mask2, then the size of nested_tiflist[group] should always be 4. 
            if len(nested_tiflist[group])==4:
                tmask=tf.TiffFile(nested_tiflist[group][3][image]).asarray()
                tmasknew=imresize(tmask,size=(x,y),interp="bilinear")
                Mask2[image,:,:]=label(tmasknew)
                #tlabel=label(tf.TiffFile(nested_tiflist[group][3][image]).asarray())
                #Mask2[image,:,:]=imresize(tlabel,size=(x,y),interp="bilinear")

            #takes the and of Mask1 and Mask2 in the event they are significantly different
            Mask=Mask1*Mask2

            #loops through labeled masks and sums up the number of unique regions in the mask, these get appended to
            #a list since they are variable in size. 
            region_num=[]
            for r in range(0,int(np.max(Mask[image,:,:]))):
                p=np.sum(Mask[image]==(r+1))
                region_num.append(p)
            image_num.append(region_num)
            dim1=[Mask[image].shape[0],Mask[image].shape[1]]
            dim2.append(dim1)
        group_num.append(image_num)
        dim3.append(dim2)
        #adds all ndimage arrays into a list
        tif_container.append([F1,F2,Mask])

    #tif_container[exp groups][F1,F2,Mask][img,ch,x,y] or [img,x,y](for Mask)    
    self.dim=dim3
    self.shape=group_num
    self.tif_array=tif_container
```

## Sort Image Data

This step sorts only the masked pixels and labels each pixel spectra with which group, image, and region it is part of along with its x and y location. Since the masked regions are typically small compared to the whole image this greatly reduces the file size and memory usage because all of the pixels outside of the mask dont need to be saved.
