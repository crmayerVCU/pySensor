```python
def calc_gamma(D1,D2,A1,A2):
        
        """Calculates gamma as specified in Arsenovic, Paul T., Carl R. Mayer, and Daniel E. Conway. "SensorFRET:
        A Standardless Approach to Measuring Pixel-based Spectral Bleed-through and FRET Efficiency using Spectral Imaging."
        Scientific reports 7.1 (2017): 15609.
        
        get paper at https://www.nature.com/articles/s41598-017-15411-8"""
        
        #Extinction Coefficient ratio product, named Gamma in article quoted above
        gamma=D2/D1*A1/A2
        return(gamma)

def gamma_test(calib_object,observation="region"):
    
    if observation=="pixel":
        observation="pix"
    

    locations_id=observation+"_locations"

    object_id=observation+"_obj"

    locations_array=getattr(calib_object,locations_id)

    #Unpack the obs ids
    group,image,region=locations_array[:,0],locations_array[:,1],locations_array[:,2]
    x,y,area=locations_array[:,3],locations_array[:,5],locations_array[:,5]

    #select donor or acceptor locations
    donor_screen=group==calib_object.group_index_donor
    acceptor_screen=group==calib_object.group_index_acceptor

    Donor_list=[]
    Acceptor_list=[]
    object_array=getattr(calib_object,object_id)
    for obs,d,a in zip(object_array,donor_screen,acceptor_screen):
        if d==True:
            Donor_list.append(obs)
        if a==True:
            Acceptor_list.append(obs)

    donor_N=len(Donor_list)
    acceptor_N=len(Acceptor_list)

    gamma_N=donor_N*acceptor_N

    if gamma_N>1e6:
        #determined empirically in testing
        calcs_per_second=715520.
        estimated_seconds=gamma_N/calcs_per_second
        estimated_minutes=estimated_seconds/60.
        
        proceed=raw_input(str(gamma_N)+" gamma calculations required. This will take an estimated "+str(estimated_minutes)+
                          " minutes.\n Do you want to proceed? (y=1,n=0)")

        if int(proceed)==0:
            return 


    gamma=[]
    for i in range(len(Donor_list)):
        for j in range(len(Acceptor_list)):
            D=Donor_list[i]
            A=Acceptor_list[j]
            #print(D.D1, D.D2, A.A1, A.A2)
            #print(C.calc_gamma(D.D1, D.D2, A.A1, A.A2))
            gamma.append(calc_gamma(D.D1, D.D2, A.A1, A.A2))
    gamma=np.array(gamma)
    return(gamma)


#% time gamma_test(C,observation="region")

gamma_vals=gamma_test(C,observation="region")
```
