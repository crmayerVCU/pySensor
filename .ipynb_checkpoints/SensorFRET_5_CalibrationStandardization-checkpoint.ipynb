## Define Custom Functions

```python

#functions for reading file heirarchy and importing image data

def get_tifname(path):
    l=[]
    for (dirname,dirs,files) in os.walk(path):
        for filename in files:
            if filename.endswith(".tif"):
                #print(filename)
                l.append(os.path.join(path,filename))
    return(l)

def get_tif_container(tifnamelist):
    container=[]
    for image in range(len(tifnamelist)):
        container.append(tf.tifffile.imread(tifnamelist[image]))
    container=np.array(container)
    return(container)

def get_spectra(tifcontainer,mask_container):
    spec_list=[]
    for image in range(tifcontainer.shape[0]):
        spec=np.zeros(tifcontainer.shape[1])
        for freq in range(tifcontainer.shape[1]):
            spec[freq]=np.nanmean(tifcontainer[image][freq][mask_container[image]])
        spec_list.append(spec)
    spec_list=np.array(spec_list)
    return(spec_list)


#functions for determining gamma and the components for the calibration

def rmbg(spectra):
    return(spectra-spectra.min())

def norm_spectra(spectra):
    spectra=rmbg(spectra)
    return(spectra/trapz(spectra))

def get_gamma(D1ave,D2ave,A1ave,A2ave,comps1,comps2):
    D2ave=rmbg(D2ave)
    D1ave=rmbg(D1ave)
    A1ave=rmbg(A1ave)
    A2ave=rmbg(A2ave)

    D2,DAF2=nnls(np.array([comps2[0],comps2[2]]).T,D2ave)[0][:]
    D1,DAF1=nnls(np.array([comps1[0],comps1[2]]).T,D1ave)[0][:]
    A1,AAF1=nnls(np.array([comps2[1],comps2[2]]).T,A1ave)[0][:]
    A2,AAF2=nnls(np.array([comps1[1],comps1[2]]).T,A2ave)[0][:]
    
    gamma=D2/D1*A1/A2
    
    return(gamma,[D1,D2,A1,A2])
        
def plot_gamma(D1ave,D2ave,A1ave,A2ave,comps1,comps2,Drange=[0,1],Arange=[0,5]):
    D2ave=rmbg(D2ave)
    D1ave=rmbg(D1ave)
    A1ave=rmbg(A1ave)
    A2ave=rmbg(A2ave)

    D2,DAF2=nnls(np.array([comps2[0],comps2[2]]).T,D2ave)[0][:]
    D1,DAF1=nnls(np.array([comps1[0],comps1[2]]).T,D1ave)[0][:]
    A1,AAF1=nnls(np.array([comps2[1],comps2[2]]).T,A1ave)[0][:]
    A2,AAF2=nnls(np.array([comps1[1],comps1[2]]).T,A2ave)[0][:]
    
    #fitting coefficients ratios
    D21calc=D2/D1
    A12calc=A1/A2
        
    #plot
    D21=(D2ave-DAF2*comps2[2])/(D1ave-DAF1*comps1[2])
    A12=(A1ave-AAF1*comps1[2])/(A2ave-AAF2*comps2[2])
    
    plot(wl_em,D21,label=D_name+' Ratio',color="blue")
    axhline(D21calc,label=D_name+' fit',ls="--",color="blue")
    ylim(Drange)
    ylabel(D_name+', '+ex2_name+'nm /'+ex1_name+'nm Ratio')
    legend(loc=2)
    xlabel("Wavelengths[nm]")
    
    twinx()
    plot(wl_em,A12,label=A_name+' Ratio',color="red")
    axhline(A12calc,label=A_name+' fit',ls="--",color="red")
    ylim(Arange)
    ylabel(A_name+', '+ex1_name+'nm /'+ex2_name+'nm Ratio')
    legend()
    
    fig, (sub1,sub2)=subplots(1,2,figsize=(14,7))
    sub2.plot(wl_em,A1ave,label=A_name+' '+ex1_name+'nm fit',color="blue")
    sub2.plot(wl_em,A2ave,label=A_name+' '+ex2_name+'nm',color="red")
    sub2.plot(wl_em,A1*comps1[1],label=A_name+' '+ex1_name+'nm fit',ls="--",color="blue")
    sub2.plot(wl_em,A2*comps2[1],label=A_name+' '+ex2_name+'nm fit',ls="--",color="red")
    sub2.plot(wl_em,AAF1*comps1[2],label='AF '+ex1_name+'nm fit',ls=":",color="blue")
    sub2.plot(wl_em,AAF2*comps2[2],label='AF '+ex2_name+'nm fit',ls=":",color="red")
    sub2.set_xlabel("Wavelengths[nm]")
    sub2.set_ylabel("Intensity[Au]")
    sub2.legend()
    
    sub1.plot(wl_em,D1ave,label=D_name+' '+ex1_name+'nm',color="blue")
    sub1.plot(wl_em,D2ave,label=D_name+' '+ex2_name+'nm',color="red")
    sub1.plot(wl_em,D1*comps1[0],label=D_name+' '+ex1_name+'nm fit',ls="--",color="blue")
    sub1.plot(wl_em,D2*comps2[0],label=D_name+' '+ex2_name+'nm fit',ls="--",color="red")
    sub1.plot(wl_em,DAF1*comps1[2],label='AF '+ex1_name+'nm fit',ls=":",color="blue")
    sub1.plot(wl_em,DAF2*comps2[2],label='AF '+ex2_name+'nm fit',ls=":",color="red")
    sub1.set_xlabel("Wavelengths[nm]")
    sub1.set_ylabel("Intensity[Au]")
    sub1.legend()  


#functions for residual/error simulations

def addnoise(F,m,b):
    Fn=F+(m*F**.5+b)*np.random.normal(0,1,F.shape)
    return(Fn)
def unmix(comps,raw):
    Mag,res=nnls(comps,raw)
    if any(Mag<.01*np.mean(Mag)):
        coefs,res=nnls(np.hstack([comps,comps*-1]),raw)
        Mag=np.array([coefs[0]-coefs[2],coefs[1]-coefs[3]])
    return(Mag,res)
def calcE(F1,F2,comps1,comps2,gamma,Qd,Qa):
    coefs1, r1 = unmix(comps1.T,F1)
    coefs2, r2 = unmix(comps2.T,F2)
    Res1=r1/trapz(F1)        
    Res2=r2/trapz(F2)

    D1=coefs1[0]
    D2=coefs2[0]
    A1=coefs1[1]
    A2=coefs2[1]

    alpha=D2/D1
    beta=A2-alpha*A1
    A1_dir=beta/(alpha*(gamma**-1-1))
    A2_dir=beta/(1-gamma)                    
    Eff1=(((D1*Qa)/((A1-A1_dir)*Qd))+1)**-1
    Eff2=(((D2*Qa)/((A2-A2_dir)*Qd))+1)**-1
    return(Eff1,Eff2,Res1,Res2,r1,r2,trapz(F1),trapz(F2))

def ResCI(Res,E,xbins,ybins,plot=0):

    H,Xcoor,Ycoor=histogram2d(Res,E,bins=[xbins,ybins],normed=True)
    CI=np.zeros([4,Xcoor.shape[0]])
    dist=np.zeros(H.shape)
    std_high=.5+(.682/2)
    std_low=.5-(.682/2)
    for i in range (0,Xcoor[:-1].shape[0]):
        
        dist[i,:]=cumsum(H[i,:]/cumsum(H[i,:])[-1])
        lb=np.interp(std_low,dist[i,:],Ycoor[:-1])
        med=np.interp(.5,dist[i,:],Ycoor[:-1])
        #m2=np.mean()
        ub=np.interp(std_high,dist[i,:],Ycoor[:-1])
        CI[:,i]=[Xcoor[i],med,lb,ub]

    if plot==1:
        CIfig, sub1=subplots(1,1)
        sub1.plot(Res,E,marker='o',alpha=.1,ls='',color='cyan',label='Simulated Data')
        sub1.plot(CI[0,:-1],CI[1,:-1],label='Mean',color='blue')
        sub1.plot(CI[0,:-1],CI[2,:-1],label='+/- 1 Standard Deviation',color='red',ls='--')
        sub1.plot(CI[0,:-1],CI[3,:-1],color='red',ls='--')
        sub1.set_xlim(xbins[0],xbins[-1])
        sub1.set_ylim(-.2,1.2)
        sub1.set_xlabel('Normalized Fit Residual')
        sub1.set_ylabel('FRET Efficiency')
        sub1.legend(loc=2)
        sub1.set_xscale('log')

    return(CI,dist)

print('Done!')
```

## Calculate Efficiency from FRET spectra

```python
F1_coefs,res1=unmix(comps1.T,F1ave)
F2_coefs,res2=unmix(comps2.T,F2ave)

F1_D=F1_coefs[0]
F1_A=F1_coefs[1]
F2_D=F2_coefs[0]
F2_A=F2_coefs[1]
R1=res1/trapz(F1ave)
R2=res2/trapz(F2ave)

alpha=F2_D/F1_D
beta=F2_A-alpha*F1_A
F2_A_dir=beta/(1-gamma)
F1_A_dir=beta/(alpha*(gamma**-1-1))

F1corr=F1ave-F1_A_dir*comps1[1]
F2corr=F2ave-F2_A_dir*comps2[1]

Eff1=(((F1_D*Qa)/((F1_A-F1_A_dir)*Qd))+1)**-1
Eff2=(((F2_D*Qa)/((F2_A-F2_A_dir)*Qd))+1)**-1

print('FRET Efficiency: '+str(Eff2)+','+str(Eff1))
print('Normalized Fit Residual: '+str(R1)+','+str(R2))
```

```python
#number of simulated pixels
pix=100    

#Simulated Efficiency Values
E_arr=np.linspace(0,1,11)

#inverse intensities - changes the signal to noise ratio in the simulation
I1=65000/np.max((Dex1*((1.0-1)*comps1[0]*Qd+1*comps1[1]*Qa)+Aex1*comps1[1]*Qa))
I2=65000/np.max((Dex2*((1.0-1)*comps2[0]*Qd+1*comps2[1]*Qa)+Aex2*comps2[1]*Qa))
Ints=np.logspace(-1,3,150)

#noise parameters
m=16.4
b=171.8


FRET_sim=np.zeros([E_arr.shape[0],2,channels-len(alist)])
fig,(sub1,sub2)=subplots(1,2)

for e in range(E_arr.shape[0]):
    F1t=(Dex1*((1.0-E_arr[e])*comps1[0]*Qd+E_arr[e]*comps1[1]*Qa)+Aex1*comps1[1]*Qa)
    F2t=(Dex2*((1.0-E_arr[e])*comps2[0]*Qd+E_arr[e]*comps2[1]*Qa)+Aex2*comps2[1]*Qa)
    FRET_sim[e,0,:]=F1t*I1
    FRET_sim[e,1,:]=F2t*I2
    sub1.plot(wl_em,FRET_sim[e,0,:],color=[E_arr[e],0,1-E_arr[e]],label='FRET: '+str(E_arr[e]))
    sub2.plot(wl_em,FRET_sim[e,1,:],color=[E_arr[e],0,1-E_arr[e]],label='FRET: '+str(E_arr[e]))
sub1.legend()
sub1.set_title(ex1_name)
sub1.set_xlabel('Emission Wavelength (nm)')
sub1.set_ylabel('Intensity (au)')
sub2.legend()
sub2.set_title(ex2_name)
sub2.set_xlabel('Emission Wavelength (nm)')
sub2.set_ylabel('Intensity (au)')

FRET_noise=np.zeros([pix,Ints.shape[0],E_arr.shape[0],2,channels-len(alist)])
comp_sim=np.zeros([Ints.shape[0],2,2,channels-len(alist)])
for i in range (Ints.shape[0]):
    comp_sim[i,0,0,:]=I1*Dex1*comps1[0]*Qd*Ints[i]
    comp_sim[i,0,1,:]=I2*Dex2*comps2[0]*Qd*Ints[i]
    comp_sim[i,1,0,:]=I1*Aex1*comps1[1]*Qa*Ints[i]
    comp_sim[i,1,1,:]=I2*Aex2*comps2[1]*Qa*Ints[i]
    for p in range (pix):
        FRET_noise[p,i,:,:,:]=addnoise(FRET_sim[:,:,:]*Ints[i],m,b)
        
```

```python
def LuxFRET(F1,F2,D1,D2,A1,A2,Qd,Qa):
    Dem=(D1+D2)/trapz(D1+D2)
    Aem=(A1+A2)/trapz(A1+A2)
    r1=(unmix(np.array([Dem,Aem]).T,D1)[0][0]/unmix(np.array([Dem,Aem]).T,A1)[0][1])*Qa/Qd
    r2=(unmix(np.array([Dem,Aem]).T,D2)[0][0]/unmix(np.array([Dem,Aem]).T,A2)[0][1])*Qa/Qd
    #X1=(D1*Qa*Aem)/(A1*Qd*Dem)*1.0
    #X2=(D2*Qa*Aem)/(A2*Qd*Dem)*1.0
    #r1=np.mean(X1[((D1*A1)>(np.amax(.02*D1*A1)))])
    #r2=np.mean(X2[((D2*A2)>(np.amax(.02*D2*A2)))])
    #print(r1,r2)
    
    d1,a1=nnls(np.array([D1,A1]).T,F1)[0]
    d2,a2=nnls(np.array([D2,A2]).T,F2)[0]

    dr=r2-r1
    da=a2-a1
    Rt=(a1*r2-a2*r1)/(dr*d1+da)
    Rt=1

    Efd=(da)/(dr*d1+da)
    Efa=(Rt*da)/(a1*r2-a2*r1)
    return (Efd,Efa,Rt)

def sRET (F1,F2,D1,D2,A1,A2,Qd,Qa):
    k1=trapz(D1)/trapz(A1)
    k2=trapz(D2)/trapz(A2)
    dk=(k2-k1)

    components1=array([D1,A1]).T
    components2=array([D2,A2]).T
    coefs1=nnls(components1,F1)[0]
    coefs2=nnls(components2,F2)[0]
    d1=trapz(coefs1[0]*D1/trapz(D1))
    d2=trapz(coefs2[0]*D2/trapz(D2))
    a1=trapz(coefs1[1]*A1/trapz(A1))
    a2=trapz(coefs2[1]*A2/trapz(A2))

    dA=a2-a1
    QD1=d1*(Qa/Qd)
    QD2=d2*(Qa/Qd)

    E1=(dA)/(dA+(QD1*dk))
    E2=(dA)/(dA+(QD2*dk))
    return (E1,E2)

```
